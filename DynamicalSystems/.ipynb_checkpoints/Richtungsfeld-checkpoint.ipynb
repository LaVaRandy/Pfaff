```python
import matplotlib.pyplot as plt
from scipy import *
from scipy import integrate
from scipy.integrate import ode
import numpy as np
global A
```

```python
## Vector field function
def vf(x,t=0):
    global A
    return (A[0][0]*x[0] + A[0][1]*x[1], A[1][0]*x[0] + A[1][1]*x[1])
```

```python
A = [[-.2, -1],[1,0]]
```

```python
#sns.set()
#plt.xkcd()
fig = plt.figure(num=1)
ax=fig.add_subplot(111)

#Vector field
X1,X2 = np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20) )
DX1, DX2 = vf([X1,X2])
#Normalize arrows
N = np.sqrt(DX1**2+DX2**2)  
DX1N, DX2N = DX1/N, DX2/N
ax.quiver(X1,X2,DX1,DX2)
t = np.linspace(0,100,10000)
P = integrate.odeint(vf, [3,3], t).T
ax.plot(P[0],P[1])
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.xlabel(r"$x1$")
plt.ylabel(r"$x2$")
plt.savefig('DF' + str(A) + '-XKCD.png')
```
