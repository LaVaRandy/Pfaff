```python
import matplotlib.pyplot as plt
from scipy import integrate
import numpy as np
global A
```

```python
## Vector field function
def vf(x,t=0):
    global A
    return (A[0][0]*x[0] + A[0][1]*x[1], A[1][0]*x[0] + A[1][1]*x[1])
```

```python
## Vector field function
def vf(x,t=0):
    mu = 0
    return [x[1], -1*np.sin(x[0]) - mu*x[1]]
```

```python
fig = plt.figure(num=1, figsize = (12,5))
ax=fig.add_subplot(111)

t = np.linspace(0,50,1000)

X0 = [np.array([1,0]), np.array([-10,1]), np.array([3.2,0]), np.array([3.1,0])]

#Vector field
X1,X2 = np.meshgrid(np.linspace(-4*np.pi,4*np.pi,40),np.linspace(-np.pi/2,np.pi/2,10))
DX1, DX2 = vf([X1,X2])
#Normalize arrows
N = np.sqrt(DX1**2+DX2**2)  
DX1N, DX2N = DX1/N, DX2/N
ax.quiver(X1,X2,DX1,DX2)
for x0 in X0:
    P = integrate.odeint(vf, x0, t).T
    ax.plot(P[0],P[1], label = '$x_0$='+str(x0))
plt.xlim([-4*np.pi, 4*np.pi])
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
```
